```python
missing_values_market_df = round(market_df.isna().sum() / market_df.shape[0], 2)
high_empty_values_columns = []
for column_name, column_value in missing_values_market_df.items():
    if column_value >= 0.2:
        high_empty_values_columns.append(column_name)
```

```python
market_df = market_df.drop(columns='Unnamed: 0')
market_df = market_df.drop(columns=high_empty_values_columns)
```
